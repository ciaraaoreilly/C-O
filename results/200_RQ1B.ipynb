## 1.
# Total Accidents

The below graph shows the % of accidents which occurred during the day and night for all recorded accidents from 1979 - 2016. 

```python
all_day_night_severity['Total'] = all_day_night_severity.sum(axis=1) #create a total column with sum of all row values
all_day_night_severity['Total'].plot(kind='pie', autopct='%1.1f%%', colors=['cornflowerblue', 'indigo'], shadow=True, figsize=(12,7))
plt.title('% of Accidents in Day vs Night, 1979 - 2016', fontsize=30) #add title
plt.axis('off') #no axis
```

We can see very clearly from this pie chart that there are far more accidents during the day time than during the night time. However, we must normalise this data to account for the fact that more people use the roads during the day than during the night time.

## 2.
# Accident Severity 

The below graph shows the normalised % of fatal, serious and slight accident outcomes during the day and night for all recorded accidents from 1979 - 2016.

```python
norm_day_night_severity.plot(kind='bar', figsize=(12,8)).legend(bbox_to_anchor=(1,1))
plt.title("Accident Severity during Day/Night", fontsize=30)
plt.xlabel("Day/Night")
plt.ylabel("% of Accidents")
```

This graph shows the ranking of accident severity for both day and night as follows:

   **Day**   -------------------------------------->     **Night**

1. slight (~82%)     -------------->   1. slight(~77%)
2. serious (~16%)    ------------>   2. serious (~20%)
3. fatal (~1%)       ------------------>   3. fatal (~2%)

So, this shows us that accidents which occur during the Night have more serious and fatal outcomes than accidents which occur during the day. The number of slight outcomes during the night is also less than that of accidents during the day time. This perhaps shows that night-time accidents are more dangerous/injurious than day-time accidents.

## 3.
# Accidents by Day of Week

The graph below shows the % of Road Accidents which occur at day time and night time for each day of the week

```python
norm_all_by_day.plot(kind='bar', figsize=(12,8), color=['cornflowerblue', 'lime', 'yellow', 'indigo', 'magenta', 'lightblue', 'orange']).legend(bbox_to_anchor=(1,1))
plt.title("% of Accidents Monday to Sunday, 1979 - 2016", fontsize=30)
plt.xlabel("Day/Night")
plt.ylabel("% of Accidents")
```

This graph gives us an interesting breakdown of the number of accidents which have occurred during the Day and Night for each day of the week. We can see;

* Saturday tops the graph with the highest percentage of accidents occurring in both day and night.
* There are more accidents at night time on a Sunday than Day time, which may blow the myth of the 'Sunday driver' idea, or may not...
* Monday has the lowest recorded % of Day time Road Accidents
* Tuesday has the lowest recorded % of Night time Road Accidents

Although we have normalised this data, these results may be influenced by the high volume of people using the roads on Saturdays. (eg. due to being off work)

However, we must also take into account that there are a lot of drivers using the roads in the morning and evenings during the week, to get to and from work, and this will also have an effect on our results.


## 4.
# Conclusions

## It _is_ safer to drive in the Day-time than in the Night-time

The graphs above have shown to us that although more accidents occur during the day, this is mainly due to the volume of people using the roads. The deciding factor in our conclusion was the accident severity breakdown of accidents which occurred during the day time and the night time. These results have shown us that accidents which occur during the night time, though fewer, have more fatal and serious outcomes. 

So, although you may be more likely to be involved in a road accident during the day time, you have a better chance of surviving with only slight injuries, than you would at night.

## 1.
# Sex of Driver

```python
norm_num_by_purp_sex.T.plot(kind='pie', subplots=True, autopct='%1.1f%%', figsize=(12,8), colors=['lightpink','cornflowerblue'], legend=False)
```

This plot shows the percentage of Male and Female drivers who were involved in road accidents, who's purpose of journey was either occupational or non-occupational at the time of the accident.

It is important to note that this ratio of male to female drivers is to be expected, as more males drive than females as a whole in the UK.

It is however interesting to note that there is a higher imbalance of male to female drivers who's purpose of journey was occupational. This perhaps indicates that more Males are employed in occupations which require driving.


## 2.
# Age Band of Driver

The below plot represents the % of drivers who were involved in road accidents. Included in the graph is the journey purpose of driver and the age range of driver. It is also important to note that this data has been normalised.

```python
norm_num_by_purp_age.T.plot(kind='bar', figsize=(12,8)).legend(bbox_to_anchor=(1,1)) #plot a bar chart
plt.xlabel('Age Band of Driver') #include axis labels
plt.ylabel('% of Accidents')
plt.title('Occupational v Recreational, 2005 - 2014', fontsize=30) #adding plot title
```

This graph clearly displays the proportion of occupational and non-occupational drivers per age group, who were involved in road accidents. Most notably, we can see that drivers between the ages of 26 - 75 were more commonly driving occupationally.


## 3.
# Conclusion

From our research into this topic, we have concluded drivers who were involved in accidents from 2005 - 2014, between the ages of 26 - 75, who's purpose of journey was occupational, were involved in more road accidents than those with a non-occupational purpose of journey, of the same age.

However, we can not conclusively say that 'occupational drivers' are more dangerous than 'non-occupational' drivers. This is due to the fact that those who drive as part of their occupation are using the roads for much longer periods of time than others. We must also take into account that the legal age for driving in the UK is 17. Both of these factors would have a part to play in our data, and as such we feel that further research into these aspects could be an interestic topic of future work.
